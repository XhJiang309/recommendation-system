## DATA PRE-PROCESSING

For building the movie recommender system we only need to analyse the following attributes of a movie in our dataset: Name of the movie,Overview,Genre,actors and directors.

```python
#We only need 6 columns 
df = df[['id','Movie_Name','overview','Genres','cast','crew']]
df.head()
df['cast'][0]
#df.shape
```

```python
#Lets check for null values for the above columns
df.isnull().sum()
```

Here we observe that 64 movies don't have overview. So we remove these rows from our dataset.

```python
#Removing all rows with no overviews
df.dropna(inplace=True)
df.shape
```

```python
#Removing strings from strings of list to change each of the following features into a list
import ast
feature =['Genres','cast','crew']
for f in feature:
    df[f]=df[f].apply(ast.literal_eval)
```

```python
#Function to extract four actors of a movie
def cast_extract(text):
    L = []
    cntr = 0
    for i in text:
        if cntr <4:
            L.append(i['name'])
        cntr+=1
    return L 
```

```python
df['cast'][0]
```

```python
df['cast'] = df['cast'].apply(cast_extract)
df['cast'].head(5)
```

```python
df['crew'][0]
```

```python
#Function to extract director of a movie
def director_extract(text):
    L = []
    for i in text:
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
df['crew'] = df['crew'].apply(director_extract)
df['crew'].head(5)
```

```python
#Function to extract genres of moive
def genre_extract(text):
    L = []
    for i in text:
        L.append(i['name']) 
    return L 
```

```python
df['Genres'][0]
```

```python
df['Genres'] = df['Genres'].apply(genre_extract)
df['Genres'].head(5)
```

```python
#Removing gap between actors,director names and genre words(if any)
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
df['Genres'] = df['Genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

We removed the gap between actor names,director names, genres.Why?
Lets understand it with the help of an example. For instance, consider the name Sandra Bullock. 
Sandra Bullock should indicate name of a single individual.However, when we take spaces into account,it points to two different entities while doing textual analysis.
Secondly, suppose we have Sandra Smith as another actor.Now our recommender system will fail to distinguish between the two actors as both have same first names. So we may like movies of Sandra Bullock but it could recommend us movies of Sandra Smith. Hence we remove the spaces between their names to preserve their unique identity.

```python
df.head()
```

```python
#Changing string to list for overview of a movie
df['overview'] = df['overview'].apply(lambda x:x.split())
df.head(5)
#We changed the overview from string to list so that we can concantenate it with rest of the columns as done in next step.
```

```python
df['movie_content'] = df['overview'] + df['Genres'] + df['cast'] + df['crew']
```

```python
df4 = df[['id','Movie_Name','movie_content']]
```

```python
df4['movie_content']
```

In order to reduce the word count, we should reduce the all the letters to lower case. For doing so, we first convert each of these lists into respective strings.

```python
#Changing list to string
df4['movie_content'] = df4['movie_content'].apply(lambda x: " ".join(x))
df4.head()
```

```python
A =['abc','defi','pqr']
k = " ".join(A)
k.split()
```

```python
#Changing all letters to lower case to reduce word count
df4['movie_content']=df4['movie_content'].apply(lambda x:x.lower())
df4['movie_content']
```

```python
from nltk.stem.porter import PorterStemmer
port=PorterStemmer()
```

Now we want to further reduce our word count as words such as like,liked,liking and so on will be treated as different words. So we should change the words to its origin such as like here.

```python
#Changing words to its roots and removing repeation of common words in order to reduce word count
def stem_fun(text):
    a=[]
    for i in text.split():
        a.append(port.stem(i))
    return " " .join(a) 
```

```python
df4['movie_content']=df4['movie_content'].apply(stem_fun)
df4['movie_content']
```

```python
#Now we are changing text(movie_content) to numeric vectors by Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df4['movie_content']).toarray()
vector.shape
```

```python
cv.get_feature_names()
#These are most frequently used words in movie content.
```

## Building the (content based)recommender system

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#To calculate similarity scores of movies
similar_movies = cosine_similarity(vector)
```

```python
similar_movies
#Here we check how similar movies are to each other by computing similarity for each of the rows.
```

```python
#Recommender system function
def recommender(movie):
    index = df4[df4['Movie_Name'] == movie].index[0]
    distances = sorted(list(enumerate(similar_movies[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df4.iloc[i[0]].Movie_Name)
```

```python
#Lets try!
recommender('The Northman')
```
